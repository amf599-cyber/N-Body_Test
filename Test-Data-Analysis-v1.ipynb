```python
import pandas as pd

# Load the dataset
df = pd.read_csv('particle_accelerations.csv')

# Display the first few rows and column names to verify headers
print("DataFrame Columns:", df.columns.tolist())
print("\nFirst 5 rows of the DataFrame:\n")
print(df.head())
```

```python
# Calculate the sum of accX, accY, and accZ components per iteration
summed_components_per_iteration = df.groupby('iteration')[['accX', 'accY', 'accZ']].sum()

# Calculate the magnitude of the resultant vector for each iteration
resultant_magnitude_per_iteration = (summed_components_per_iteration['accX']**2 +
                                     summed_components_per_iteration['accY']**2 +
                                     summed_components_per_iteration['accZ']**2)**0.5

print("Resultant magnitude of acceleration per iteration (first 5 rows):\n")
print(resultant_magnitude_per_iteration.head())
```

```python
# Calculate total magnitude of acceleration per iteration
total_acceleration_per_iteration = df.groupby('iteration')['magnitude'].sum()
print("Total acceleration magnitude per iteration (first 5 rows):\n")
print(total_acceleration_per_iteration.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# Determine the number of iterations for the first n%
num_iterations = len(total_acceleration_per_iteration)
first_10_percent_iterations = int(num_iterations * 0.001)

# Filter the data to include only the first n% of iterations
filtered_total_acceleration = total_acceleration_per_iteration.head(first_10_percent_iterations)
filtered_resultant_magnitude = resultant_magnitude_per_iteration.head(first_10_percent_iterations)

# Plot 1: Total Acceleration Magnitude Over Iterations (sum of individual particle magnitudes) for the first n%
plt.figure(figsize=(12, 6))
plt.plot(filtered_total_acceleration.index, filtered_total_acceleration.values, label='Sum of Individual Magnitudes')
plt.title('Total Acceleration Magnitude (Sum of Individual Magnitudes) Over First 0.010% Iterations')
plt.xlabel('Iteration')
plt.ylabel('Total Acceleration Magnitude')
plt.legend()
plt.grid(True)
plt.show()

# Plot 2: Resultant Magnitude of Acceleration Over Iterations (magnitude of net acceleration vector) for the first n%
plt.figure(figsize=(12, 6))
plt.plot(filtered_resultant_magnitude.index, filtered_resultant_magnitude.values, color='green', label='Magnitude of Net Acceleration Vector')
plt.title('Resultant Magnitude of Acceleration (Magnitude of Net Vector) Over First 0.010% Iterations')
plt.xlabel('Iteration')
plt.ylabel('Resultant Acceleration Magnitude')
plt.legend()
plt.grid(True)
plt.show()
```
